### Step 3: Defining objects and building lists

```python
mrp_list = []
price_list = []
offer_list = []
product_list = []
rating_list = []
reviews_list = []
features_list = []
for i in range(1,41):
    URL = 'https://www.flipkart.com/search?q=mobiles&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off&as-pos=1&as-type=HISTORY&page='+str(i)
    page = requests.get(URL)
    pagecontent = page.text
    soup = BeautifulSoup(pagecontent)
    list_1 = soup.find_all('div',attrs={'class':'_3pLy-c row'})
    
    for j in list_1:
        mrp = j.find('div',attrs={'class':'_3I9_wc _27UcVY'})
        if mrp is None:
            mrp_list.append(np.NaN)
        else:    
            print(mrp.text)
            mrp_list.append(mrp.text)
            
        price = j.find('div',attrs={'class':'_30jeq3 _1_WHN1'})
        if price is None:
            price_list.append(np.NaN)
        else:    
            print(price.text)
            price_list.append(price.text)
        
        offer = j.find('div',attrs={'class':'_3Ay6Sb'})
        if offer is None:
            offer_list.append(np.NaN)
        else:    
            print(offer.text)
            offer_list.append(offer.text)
            
        product = j.find('div',attrs={'class':'_4rR01T'})
        if product is None:
            product_list.append(np.NaN)
        else:    
            print(product.text)
            product_list.append(product.text)
           
        rating = j.find('div',attrs={'class':'_3LWZlK'})
        if rating is None:
            rating_list.append(np.NaN)
        else:
            print(rating.text)
            rating_list.append(rating.text)
            
        reviews = j.find('span',attrs={'class':'_2_R_DZ'})
        if reviews is None:
            reviews_list.append(np.NaN)
        else:    
            print(reviews.text)
            reviews_list.append(reviews.text)
            
        features = j.find('ul',attrs={'class':'_1xgFaf'})    
        if features is None:
            features_list.append(np.NaN)
        else:    
            print(features.text)
            features_list.append(features.text)
        
```

```python
len(price_list)
```

### Step 4: Store the data in a required format

```python
mobiles = pd.DataFrame({'MRP':mrp_list,'ProductName':product_list,'Price':price_list,'Offer':offer_list,'Rating':rating_list,'Reviews':reviews_list,'Features':features_list})
```
